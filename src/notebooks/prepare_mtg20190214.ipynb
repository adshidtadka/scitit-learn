```python
import networkx as nx
import numpy as np
import requests
import json
import collections
import itertools
%matplotlib inline
import matplotlib.pyplot as plt
import h5py
```

```python
INPUT_DIR = '/home/inagaki/dev/src/10.228.155.190/brain2019/brain2019/data/dtw'
N_CONTENTS = 20
```

```python
p_id = 19
h5file = h5py.File(INPUT_DIR + '/' + "01" + "{0:02d}".format(p_id + 1) + ".hdf","r")
mat = h5file[list(h5file.keys())[0]].value[0:N_CONTENTS, 0:N_CONTENTS]
mat.shape
```

```python
content_label = [i for i in range(0, N_CONTENTS)]
fig, ax = plt.subplots(figsize=(10, 8))
heatmap = ax.pcolor(mat, cmap=plt.cm.Blues)
ax.set_xticks(np.arange(mat.shape[0]) + 0.5, minor=False)
ax.set_yticks(np.arange(mat.shape[1]) + 0.5, minor=False)
ax.invert_yaxis()
ax.xaxis.tick_top()
ax.set_xticklabels(content_label, minor=False)
ax.set_yticklabels(content_label, minor=False)
fig.colorbar(heatmap, ax=ax)
plt.show()
```
